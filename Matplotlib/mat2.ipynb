```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [10, 25, 15, 30]

plt.bar(categories, values)
plt.show()
```

```python
plt.bar(categories, values)
plt.title("Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.bar(categories, values, color=['red','blue','green','orange'])
plt.show()
```

```python
plt.barh(categories, values)
plt.show()
```

```python
import numpy as np

x = np.arange(len(categories))

values2 = [12, 20, 18, 28]

plt.bar(x - 0.2, values, width=0.4, label="2023")
plt.bar(x + 0.2, values2, width=0.4, label="2024")

plt.xticks(x, categories)
plt.legend()
plt.show()
```

```python
# scatter() plots points

# s controls size

# alpha controls transparency

# Used heavily before regression

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
plt.scatter(x, y, color='red', s=100)
plt.show()
```

```python
y2 = [1,3,4,3,4]

plt.scatter(x, y, label="Class A")
plt.scatter(x, y2, label="Class B")

plt.legend()
plt.show()
```

```python
plt.scatter(x, y, alpha=0.5)
plt.show()
```

```python
# Histogram is used for:

# Understanding distribution

# Checking skewness

# Detecting outliers

# Frequency analysis

# Key Takeaways
# hist() shows distribution
# bins controls grouping
# alpha controls transparency
# Essential for EDA


import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4,4,4,4]

plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=4)
plt.show()
```

```python
plt.hist(data, bins=4, color='blue', edgecolor='black')
plt.show()
```

```python
plt.hist(data, bins=4, alpha=0.7)
plt.show()
```

```python
data2 = [2,3,3,4,4,5,5,5]

plt.hist(data, bins=4, alpha=0.5, label="Class A")
plt.hist(data2, bins=4, alpha=0.5, label="Class B")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D"]
sizes = [30, 25, 20, 25]

plt.pie(sizes)
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Market Share")
plt.show()
```

```python
explode = [0.1, 0, 0, 0]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2)

ax[0].plot([1,2,3], [4,5,6])
ax[0].set_title("Line Plot")

ax[1].bar(["A","B","C"], [3,7,5])
ax[1].set_title("Bar Chart")

plt.show()
```

```python
# fig  pura canvas

# ax  individual graph areas

fig, ax = plt.subplots(2,2)      # tight_layout() spacing adjust

ax[0,0].plot([1,2,3], [4,5,6])
ax[0,1].scatter([1,2,3], [3,6,2])
ax[1,0].bar(["A","B","C"], [5,3,7])
ax[1,1].hist([1,2,2,3,3,3,4])

plt.tight_layout()          #tight_layout() spacing adjust
plt.show()
```

```python
# Object-Oriented Approach
# The OO approach provides full control over:
# Figures
# Axes
# Layout
# Styling


import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1,2,3], [4,5,6])
ax.set_title("Line Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

plt.show()
```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot([1,2,3], [4,5,6])
ax[0].set_title("Plot 1")

ax[1].scatter([1,2,3], [3,6,2])
ax[1].set_title("Plot 2")

plt.show()
```
