# seaborn relplot()  , figure lavel function 

| Parameter     | Meaning                           |
| ------------- | --------------------------------- |
| `data`        | Dataset (DataFrame)               |
| `x`           | Variable for x-axis               |
| `y`           | Variable for y-axis               |
| `hue`         | Color grouping variable           |
| `size`        | Marker size mapping               |
| `style`       | Marker style mapping              |
| `units`       | Identifier for repeated measures  |
| `weights`     | Weight values for observations    |
| `row`         | Facet rows                        |
| `col`         | Facet columns                     |
| `col_wrap`    | Wrap columns in grid layout       |
| `row_order`   | Order of row categories           |
| `col_order`   | Order of column categories        |
| `palette`     | Color palette                     |
| `hue_order`   | Order of hue categories           |
| `hue_norm`    | Normalize numeric hue             |
| `sizes`       | Size range                        |
| `size_order`  | Order of size categories          |
| `size_norm`   | Normalize size values             |
| `markers`     | Custom markers                    |
| `dashes`      | Line style control                |
| `style_order` | Order of style categories         |
| `legend`      | Legend control                    |
| `kind`        | `"scatter"` or `"line"`           |
| `height`      | Height of each facet              |
| `aspect`      | Aspect ratio                      |
| `facet_kws`   | Extra FacetGrid arguments         |
| `**kwargs`    | Extra arguments passed internally |




- Default graph = Scatter Plot
- kind="line" → Line Plot
- relplot() = Figure-level function
- Internally use karta hai:
      - scatterplot()
      - lineplot()

# Seaborn Summary

##  Relational Plots

| Function        | Purpose                      |
| --------------- | ---------------------------- |
| `relplot()`     | Figure-level relational plot |
| `scatterplot()` | Scatter visualization        |
| `lineplot()`    | Line visualization           |


## 2. Categorical Plots

| Function       | Purpose                  |
| -------------- | ------------------------ |
| `catplot()`    | Figure-level categorical |
| `barplot()`    | Aggregated bar chart     |
| `countplot()`  | Frequency count          |
| `boxplot()`    | Box distribution         |
| `violinplot()` | Density + box            |
| `stripplot()`  | Categorical scatter      |
| `swarmplot()`  | Non-overlapping scatter  |
| `pointplot()`  | Mean with CI             |

## 3. Distribution Plots

| Function     | Purpose                   |
| ------------ | ------------------------- |
| `displot()`  | Figure-level distribution |
| `histplot()` | Histogram                 |
| `kdeplot()`  | Kernel density            |
| `ecdfplot()` | Cumulative distribution   |
| `rugplot()`  | Data ticks                |

## 4. Matrix Plots

| Function       | Purpose                   |
| -------------- | ------------------------- |
| `heatmap()`    | Correlation visualization |
| `clustermap()` | Clustered matrix          |


## 5. Regression Plots

| Function    | Purpose                 |
| ----------- | ----------------------- |
| `regplot()` | Regression line         |
| `lmplot()`  | Figure-level regression |

## 5. Advanced Multi Plots

| Function      | Purpose                |
| ------------- | ---------------------- |
| `pairplot()`  | Pairwise relationships |
| `jointplot()` | Joint + marginal plots |
| `FacetGrid`   | Custom grid control    |


- REL → Relationship
- CAT → Category
- DIS → Distribution
- HEAT → Matrix
- REG → Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
# realtional plots :
# scatterplot()
sns.scatterplot(data=df, x="total_bill", y="tip")   # Bill vs Tip relation
```

```python
# lineplot()
sns.lineplot(data=df, x="size", y="total_bill")  # Table size vs bill trend
```

```python
# relplot()

sns.relplot(data=df, x="total_bill", y="tip", hue="sex")   # grouping supports 
```

```python
# Categorical Plots
# barplot()

sns.barplot(data=df, x="day", y="total_bill")# avg bill of every day
```

```python
# countpot() , 

sns.countplot(data=df, x="day")
```

```python
# boxplot() ,

sns.boxplot(data=df, x="day", y="total_bill")   #Distribution + outliers
```

```python
# violinplot

# Distribution shape

sns.violinplot(data=df, x="day", y="total_bill")
```

```python
# catplot()
# provide Multi-plot version 

sns.catplot(data=df, x="day", y="total_bill", kind="bar")
```

```python
# Distribution Plots

# histplot()
sns.histplot(data=df, x="total_bill", )
```

```python
sns.histplot(data=df, x="total_bill", kde=True)   # Bill distribution
```

```python
# kdplot()
# Smooth density curve


sns.kdeplot(data=df, x="total_bill")
```

```python
sns.kdeplot(data=df, x="total_bill", fill=True)
```

```python
# displot 
# Figure-level distribution


sns.displot(data=df, x="total_bill")
```

```python
# 5: Matrix Plot

# Correlation matrix

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

```python
# 6: Regression

sns.regplot(data=df, x="total_bill", y="tip")   #    Regression line
```

```python
sns.lmplot(data=df, x="total_bill", y="tip")
```

```python
# 7: Multi-variable Analysis
# pairplot()

sns.pairplot(df)   # its provide all numeric combinations 
```

```python
# jointplot()
#Scatter + side distributions


sns.jointplot(data=df, x="total_bill", y="tip", kind="scatter")
```

```python
sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    size="size",
    col="day",
    row="time",
    kind="scatter",
    palette="Set2",
    height=4,
    aspect=1.2
)
```
