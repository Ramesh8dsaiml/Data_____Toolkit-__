```python
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3, 4])
print(arr1.shape)   # Output: (4,) → 1 row with 4 elements

# 2D array
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print(arr2.shape)   # Output: (2, 3) → 2 rows and 3 columns

# 3D array
arr3 = np.array([[[1, 2], [3, 4]],[[5, 6],[7, 8]]])

print(arr3.shape)   # Output: (2, 2, 2) → 2 blocks, each with 2 rows and 2 columns

# why use>>>>> 
#It helps you understand the structure of the data.

#Required for matrix operations, reshaping, broadcasting, etc.

#Useful for debugging when shapes don't match.
```
