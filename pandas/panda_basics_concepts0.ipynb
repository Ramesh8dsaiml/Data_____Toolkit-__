```python
# # group by :
# Split â†’ Apply â†’ Combine
# Used for category-wise analytics.
# means ; split data and then calculate 

import pandas as pd

data = {
    "Department": ["CSE", "CSE", "IT", "IT", "ECE", "ECE"],
    "Student": ["A", "B", "C", "D", "E", "F"],
    "Marks": [85, 90, 78, 88, 92, 70],
    "Salary": [800000, 900000, 600000, 750000, 950000, 500000]
}

df = pd.DataFrame(data)
df
```

```python
# basics group by :
df.groupby("Department")["Marks"].mean()
```

```python
# multiple group by :
df.groupby("Department")["Marks"].agg(["mean", "max", "min", "count"])
```

```python
df.groupby("Department").agg({
    "Marks": ["mean", "max"],
    "Salary": "mean"
})
```

```python
# feature engineering  with group by 
#agg â†’ group level result
# transform â†’ original shape maintain karta hai



df["Dept_Avg"] = df.groupby("Department")["Marks"].transform("mean")
df
```

```python
df["Performance_vs_Dept"] = df["Marks"] - df["Dept_Avg"]
```

```python
df
```

```python
 # apply ( advance lavel )
# when you can apply login  than apply ( apply method )

def grade(x):
    if x >= 90:
        return "A"
    elif x >= 80:
        return "B"
    else:
        return "C"

df["Grade"] = df["Marks"].apply(grade)
```

```python
df
```

# What is Pivot Table?

- A way to summarize data into a matrix format.

- Rows â†’ index
- Columns â†’ columns
- Values â†’ aggregated metric

```python
# data summery convert into tables :
# example :
import pandas as pd

data = {
    "Department": ["CSE","CSE","IT","IT","ECE","ECE"],
    "Gender": ["Male","Female","Male","Female","Male","Female"],
    "Marks": [85,90,78,88,92,70],
    "Salary": [800000,900000,600000,750000,950000,500000]
}

df = pd.DataFrame(data)
df
```

```python
# Basic Pivot
# Department vs Gender â†’ Average Marks
pd.pivot_table(df,
               values="Marks",
               index="Department",
               columns="Gender",
               aggfunc="mean")

```

```python
# multiple agg
pd.pivot_table(df,
               values="Marks",
               index="Department",
               columns="Gender",
               aggfunc=["mean","max"])
```

```python
pd.pivot_table(df,
               values="Marks",
               index="Department",
               columns="Gender",
               aggfunc=["mean","max","min","count","median","std","sum"])
```

```python
pd.pivot_table(df,
               values="Marks",
               index="Department",
               columns="Gender",
               aggfunc="mean",
               margins=True)
```

```python
# its provide the overall avg :


pd.pivot_table(df,
               values="Marks",
               index="Department",
               columns="Gender",
               aggfunc=["mean","median","min","std","sum",],
               margins=True)
```

# Time Series (Very Important)

### date_range samajhna
2ï¸. DateTime index ka importance
3ï¸. Resampling
4ï¸. Rolling mean
5ï¸. Shift
6ï¸. Lag feature
7ï¸. Timezone
8ï¸. Timedelta

```python
import pandas as pd
import numpy as np

# Date create
dates = pd.date_range(start="2023-01-01", periods=12, freq="M")

# Random sales data
sales = np.random.randint(1000, 5000, 12)

# DataFrame create
df = pd.DataFrame({
    "Date": dates,
    "Sales": sales
})

# Date ko index bana do
df.set_index("Date", inplace=True)

df
```

```python
df.loc["2023-03"]
df.loc["2023-03":"2023-06"] # Time range select karna.
```

```python
df.resample("Q").sum()
# 

# Frequency change karna.
# Real life:
# Monthly data ko quarterly report banana.
```

```python
df["Rolling_3"] = df["Sales"].rolling(3).mean()   

# Last 3 months ka average nikalna.
# Real life:
# Trend smooth karna.
```

```python
df
```

```python
df["Previous_Month"] = df["Sales"].shift(1)
```

```python
df # Pichhle month ki value lana.
```

```python
df["Growth_%"] = df["Sales"].pct_change() * 100
```

```python
df # Month to month percentage growth.
```

```python
df["Month"] = df.index.month
df["Year"] = df.index.year
```
